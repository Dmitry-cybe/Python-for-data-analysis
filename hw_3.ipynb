```python
class1surv = (df.groupby(["Pclass"])["Survived"].sum()[1]/df['Pclass'].value_counts()[1])*100
class2surv = (df.groupby(["Pclass"])["Survived"].sum()[2]/df['Pclass'].value_counts()[2])*100
class3surv = (df.groupby(["Pclass"])["Survived"].sum()[3]/df['Pclass'].value_counts()[3])*100
print(class1surv,'% - процент выживаемости в первом классе')
print(class2surv,'% - процент выживаемости во втором классе')
print(class3surv,'% - процент выживаемости в третьем классе')
```
